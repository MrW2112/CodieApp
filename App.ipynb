```python

import re
from googletrans import Translator
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk, ImageDraw
import ttkbootstrap as ttk
from ttkbootstrap.constants import *

# Initialize the translator
translator = Translator()

def calculate(expression):
    try:
        result = eval(expression)
        return result
    except Exception as e:
        return str(e)

def translate_text(text, target_language):
    try:
        translation = translator.translate(text, dest=target_language)
        return translation.text
    except Exception as e:
        return str(e)

def translate_calculation_result(expression, target_language):
    result = calculate(expression)
    result_text = f"The Answer you are looking for {expression} is {result}"
    translated_result = translate_text(result_text, target_language)
    return translated_result

def on_translate():
    user_input = entry_input.get()
    target_language = entry_language.get()

    match = re.search(r'(\d+\.?\d*)\s*([\+\-\*/])\s*(\d+\.?\d*)', user_input)
    if match:
        num1, operator, num2 = match.groups()
        expression = f"{num1} {operator} {num2}"
        translated_result = translate_calculation_result(expression, target_language)
        messagebox.showinfo("Translation Result", translated_result)
    else:
        translation = translate_text(user_input, target_language)
        messagebox.showinfo("Translation Result", translation)

def create_rounded_image(image_path, size, radius):
    # Open the image
    image = Image.open(image_path).resize(size, Image.Resampling.LANCZOS)
    
    # Create a mask
    mask = Image.new("L", size, 0)
    draw = ImageDraw.Draw(mask)
    draw.rounded_rectangle((0, 0) + size, radius, fill=255)
    
    # Apply the mask to the image
    rounded_image = Image.new("RGBA", size)
    rounded_image.paste(image, (0, 0), mask)
    
    return rounded_image

# Create the main window
root = ttk.Window(themename="darkly")
root.title("Codie chatbot")
root.geometry("400x650")

# Create a style object
style = ttk.Style()
# Configure the style for buttons
style.configure('button_translate', font=('Arial', 25))
# Configure the style for labels
style.configure('label_language', font=('Arial', 30))

image_path = "codie.png"
# Load the image using Pillow
image = Image.open(image_path)
new_width = 300
new_height = 300
resized_image = image.resize((new_width, new_height), Image.Resampling.BICUBIC)
tk_image = ImageTk.PhotoImage(resized_image)


# Create a label to display the image
image_label = tk.Label(root, image=tk_image)
image_label.image = tk_image
image_label.pack(side=TOP, padx=20, pady=20)

# Create and place the input field
label_input = ttk.Label(root, text="Enter a calculation or text: ")
label_input.pack(pady=10)
entry_input = ttk.Entry(root)
entry_input.pack(pady=10)

# Create and place the language field
label_language = ttk.Label(root, text="Enter the desired South African language (Sesotho, Xhosa, Zulu ,Afrikaans):")
label_language.pack(pady=10)
entry_language = ttk.Entry(root)
entry_language.pack(pady=10)

# Create and place the translate button
button_translate = ttk.Button(root, text="Send", command=on_translate, bootstyle=SUCCESS)
button_translate.pack(pady=20)

# Run the application
root.mainloop()
```
